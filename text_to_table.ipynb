```python
# 'content' 열의 모든 값을 문자열로 변환
df['content'] = df['content'].astype(str)

# 명사 추출 및 빈도수 확인
text = df['content'].to_list()
nouns = []
for sentence in text:
    nouns += okt.nouns(sentence)

filtered_nouns = [noun for noun in nouns if len(noun) > 1]

count = Counter(filtered_nouns)
noun_list = count.most_common()

for noun, frequency in noun_list:
    if frequency >= 3:
        print(noun, frequency)
```

```python
fish_pattern = r'광어'
fishing_pattern = r'\b(루어낚시|찌낚시|선상낚시|원투낚시)\b'
point_pattern = r'\b(진해|목포|여수|제주도|부산|거제도|천성항|장승포|지세포|\
                     영암|송도|백야도|금오도|돌산도|신기항|일광|돌산|기장|\
                     군산|선유도|새만금|고군산군도|무녀도|장자도|우도|김해|\
                     인천|영종도|영종대교|태안|칠천도|신전항|오륙도|대변항|원전항|\
                     창원|마산합포구|가덕도|대항항|거문도|백도|북항|감천|고성|황덕도|\
                     통영|해운대|수도항|좌사리도|서귀포|\
                     향일암|영덕|덕포항|격포항|한산도|구조라항|매물도|소매물도|마검포항|\
                     서천|울릉도|방어진항|삼포항|포항|도두항|독도|\
                     완도|추자도|사계항|영흥도|자월도|오이도|가파도|마라도|외연도|욕지도|어청도|신창항|\
                     대정읍|삼천포|모슬포항|오천항|안면도|영도|무창포항|무창포방파제|대천항|비응항|시화방조제|\
                     삼길포항|장고항|당진|안흥항|서산|방포항|내포항|고흥군|보령시|속초|\
                     학암포|학암포항|대천해수욕장|대천항방파제|영목항|대부도|궁평항|덕적도|남당항|\
                     백사장포구|몽산포|몽산포항|연평도|가력도항|격렬비열도|천수만|홍원항|강릉|팔미도|삼척|국화도|\
                     부흥호|어은돌항|대난지도|드르니항|무의도|소무의도|녹도|토끼섬|왜목항|진두선착장|모항항|\
                     회변항|울산|전곡항|대야도항|주문진|승봉도|신시도|안목항|강화도|\
                     감포항|감포|장생포|가조도|왕산교|녹동항|삼목항|강구항|잠진도|탄도항|무의대교|톱머리항|거가대교|\
                     제부도|제부항|장봉도|금강대교|거잠포|오산항|창촌방파제|실미도|임원항|선재도|물치항|누에섬|\
                     후진항|대진항|황금산갯바위|진도항|영흥대교|입파도|백아도|공현진항|마량항|선촌항|굴업도|만대항|축항|백령도|울도|\
                     수산항|차귀도|양양|강릉항|대포항|다대포항|대포근포항|거진항|태종대|해금강|후포항|와현항|백야항|봉포항|묵호항|\
                     덕산항|오호항|낙산항|영진항|장승포항|애월항|이동항|민락항|사천진항|욕지항|심곡항|고현항|칠암항|삼덕항|지세포항|우동항|\
                     동백섬|상진항|전촌항|소돌항|읍천항|법환항|동산항|인구항|채석포항|궁리항|압해도|아야진항)\b'
#남항(인천/부산), 영호(고흥/고흥군), 닭섬(태안/군산/고흥/무안에 있음), 대포근포항(대포항과 근포항을 합쳐서 부르는 말/경남 거제시 남부면 저구리)
#금진항(강원도 강릉시 옥계면 헌화로/전라남도 고흥군 금산면 신촌리), 양포항(경상북도 포항시 남구 장기면 양포리/경상남도 하동군 진교면 양포리)
time_pattern = r'\b(1시|2시|3시|4시|5시|6시|7시|8시|9시|10시|11시|12시|\
                    13시|14시|15시|16시|17시|18시|19시|20시|21시|22시|23시|24시)\b'
otime_pattern = r'\b(오전|오후|새벽|낮|밤|자정|정오|아침|저녁)\b'
month_pattern = r'\b(1월|2월|3월|4월|5월|6월|7월|8월|9월|10월|11월|12월)\b'

new_data = []

# 각 문장에 대해 패턴 매칭 및 빈도수 계산 (높은 빈도수)
#locate -> 지역 구분
new_dataframe = pd.DataFrame(columns=['fish', 'fishing', 'point', 'otime', 'time', 'month'])

for i in range(len(df)):
    matches_fish = re.findall(fish_pattern, df['content'][i])
    matches_fishing = re.findall(fishing_pattern, df['content'][i])
    matches_point = re.findall(point_pattern, df['content'][i])
    matches_time = re.findall(time_pattern, df['content'][i])
    matches_otime = re.findall(otime_pattern, df['content'][i])
    matches_month = re.findall(month_pattern, df['content'][i])
    
    row_data = {}
    
    if matches_fish:
        fish_freq = Counter(matches_fish)
        most_common_fish = fish_freq.most_common(1)[0][0]
        row_data['fish'] = most_common_fish
    
    if matches_fishing:
        fishing_freq = Counter(matches_fishing)
        most_common_fishing = fishing_freq.most_common(1)[0][0]
        row_data['fishing'] = most_common_fishing
    
    if matches_point:
        point_freq = Counter(matches_point)
        most_common_point = point_freq.most_common(1)[0][0]
        row_data['point'] = most_common_point
        
    if matches_time:
        time_freq = Counter(matches_time)
        most_common_time = time_freq.most_common(1)[0][0]
        row_data['time'] = most_common_time
        
        
    if matches_otime:
        otime_freq = Counter(matches_otime)
        most_common_otime = otime_freq.most_common(1)[0][0]
        row_data['otime'] = most_common_otime
        
        
    if matches_month:
        month_freq = Counter(matches_month)
        most_common_month = month_freq.most_common(1)[0][0]
        row_data['month'] = most_common_month
    
    new_dataframe = pd.concat([new_dataframe, pd.DataFrame(row_data, index=[0])], ignore_index=True)
```
