```python
import pandas as pd
```

```python
# 블로그 글 정제 데이터 (.xlsx 파일)
blog_data = pd.read_excel(r'C:\Users\이원경\Desktop\어쩌다 어부\blog_contents_참돔_정제_4.xlsx')

# 사용자 입력 받기 (체크박스로 가정)
month = input("날짜: ")
locate = input("장소: ")
fish = input("어종: ")
fishing = input("낚시 방법: ")
time = input("시간: ")

# 추천 포인트 찾기
matched_points = []

# 데이터 일치 여부 빈도 확인
for index, row in blog_data.iterrows():
    frequency = 0
    if row['fish'] == fish:
        frequency += 1
    if row['fishing'] == fishing:
        frequency += 1
    if row['locate'] == locate:
        frequency += 1
    if row['month'] == month:
        frequency += 1
    if row['time'] == time:
        frequency += 1
        
    # 입력받은 시간 범위에 따라 데이터 일치 여부 빈도를 추가적으로 확인  
    if time == '오전 6시 ~ 정오 12시' and any(str(hour) in str(row['time']) for hour in [6, 7, 8, 9, 10, 11, 12]):
        frequency += 1
    elif time == '정오 12시 ~ 오후 18시' and any(str(hour) in str(row['time']) for hour in [12, 13, 14, 15, 16, 17, 18]):
        frequency += 1
    elif time == '오후 18시 ~ 자정 12시' and any(str(hour) in str(row['time']) for hour in [18, 19, 20, 21, 22, 23, 24]):
        frequency += 1
    elif time == '자정 12시 ~ 오전 6시' and any(str(hour) in str(row['time']) for hour in [24, 1, 2, 3, 4, 5, 6]):
        frequency += 1

    matched_points.append((row['point'], frequency))

# 데이터 일치 여부 빈도가 가장 높은 데이터 추천
if matched_points:
    max_frequency = max(matched_points, key=lambda x: x[1])[1]
    highest_frequency_points = [point[0] for point in matched_points if point[1] == max_frequency]

    # 데이터 일치 여부 빈도가 같은 데이터가 여러 개라면, 우선 순위로 판단하여 추천
    priority = ['fish', 'locate', 'fishing', 'month', 'time']
    recommended_point = None

    for attr in priority:
        for point in highest_frequency_points:
            if row[attr] == eval(attr) or eval(attr) == '':
                recommended_point = point
                break
        if recommended_point:
            break

    if recommended_point:
        print("추천 낚시 포인트:")
        print(recommended_point)
    else:
        print("해당하는 추천 낚시 포인트가 없습니다.")
else:
    print("해당하는 추천 낚시 포인트가 없습니다.")
```
